```python
import vdetect
```

```python
test_images = [mpimg.imread(f) for f in glob('test_images/*.jpg')]

main_region = vdetect.define_main_region_custom()
mr_x0, mr_y0, mr_x1, mr_y1 = main_region
```

```python
im = test_images[0]

plt.imshow(
    vdetect.draw_boxes(
        im,
        [main_region]
    )
)

print(main_region)
```
